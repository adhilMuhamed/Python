```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(5, 100, 25)
y = np.linspace(10,1000, 25)

plt.plot(x, y**2, 'ro', ms = 10)
plt.xlabel('Marketing Budget')
plt.ylabel('Revenue')
plt.title('Marketing Budget vs Revenue')
plt.show()  #display the graph
```

```python
x = np.linspace(5, 100, 25)
y = np.linspace(10,1000, 25)

plt.plot(x, y, 'ro')
plt.plot(x, y*2, 'y*')
plt.xlabel('Marketing Budget')
plt.ylabel('Revenue')
plt.title('Marketing Budget vs Revenue')
plt.show()  #display the graph
```

### Subplots

```python
x = np.linspace(5, 100, 25)
y = np.linspace(10,1000, 25)

plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.plot(x, y, 'ro')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1,2,2)
plt.plot(x, y*2, 'y*')
plt.xlabel('x')
plt.ylabel('y*2')

plt.show()
```

### Legends

```python
x = np.linspace(5, 100, 25)
y = np.linspace(10,1000, 25)

plt.plot(x, y, 'ro', label = 'x vs y')
plt.plot(x, y*2, 'y*', label = 'x vs y*2')
plt.xlabel('Marketing Budget')
plt.ylabel('Revenue')
plt.title('Marketing Budget vs Revenue')
plt.legend(loc = 'lower right', framealpha = 0.5, facecolor = 'cyan')
plt.show()  #display the graph
```

## Types of Plots

### 1. Scatter Plot

```python
data = pd.read_csv('marketing_data.csv')
data.dropna(inplace = True)
```

```python
data
```

```python
x = data['Marketing Budget (X) (In lakhs)']
y = data['Actual Sales(Y) (In crores)']

plt.scatter(x, y, s = 100, color = 'magenta', marker = '*')
plt.grid(ls = 'dashed', lw = 2, c = 'r')
plt.show()
```

```python
data.corr()
```

### 2. Pie Chart

```python
merged = pd.read_csv('merged_data.csv')
```

```python
merged.head()
```

```python
merged['Product_Category'].value_counts()
```

```python
merged['Product_Category'].unique()
```

```python
plt.pie(merged['Product_Category'].value_counts(), labels = merged['Product_Category'].unique(), 
       autopct = '%.3f', colors = ['red', 'orange', 'magenta'])
plt.show()
```

### Bar Chart

```python
plt.bar(merged['Product_Category'].unique(), height = merged['Product_Category'].value_counts(), 
        width = 0.5, color = ['magenta', 'purple', 'red'])
plt.xticks(rotation = 90)
plt.show()
```

### Scatterplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter', hue = 'sex', col = 'sex')
plt.show()
```

### Line Plot

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
sns.relplot(x = 'year', y = 'passengers', data = flights, kind = 'line')
plt.show()
```

### Box Plot(1 numerical vs 1 categorical)

```python
tips.head()
```

```python
sns.boxplot(x = 'time', y = 'total_bill', data = tips)
plt.show()
```

### Regplot(Regression Plot)

```python
tips.head()
```

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', col = 'sex')
plt.show()
```

### Distribution Plot

```python
tips.head()
```

```python
sns.kdeplot(x = 'total_bill', data = tips)
plt.show()
```

```python
sns.displot(x = 'total_bill', data = tips, kde = True)
plt.show()
```

### CountPlot

```python
tips.head()
```

```python
sns.countplot(x = 'smoker', data = tips, hue = 'sex')
plt.show()
```
