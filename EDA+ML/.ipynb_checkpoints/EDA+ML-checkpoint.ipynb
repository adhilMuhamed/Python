### Label Encoding

```python
credit_data = pd.read_excel('credit-data-training.xlsx')
```

```python
credit_data.head()
```

```python
LE = LabelEncoder()
```

```python
credit_data['Credit-Application-Result'] = LE.fit_transform(credit_data['Credit-Application-Result'])
```

```python
credit_data.head()
```

### One Hot encoding

```python
OHE = OneHotEncoder(handle_unknown = 'ignore' , sparse = False)
```

```python
OHE.fit(credit_data[['Purpose']])
```

```python
encoded = OHE.transform(credit_data[['Purpose']])
```

```python
pd.DataFrame(encoded)
```

### Mean Imputation

```python
data['Price'].fillna(data['Price'].mean())
```

### Median Imputation

```python
data['Price'].fillna(data['Price'].median())
```

### Mode imputation

```python
data
```

```python
data['Bathroom'] = data['Bathroom'].astype('object')
```

```python
data.info()
```

```python
data['Bathroom'].fillna(data['Bathroom'].mode()[0])
```

### Constant Imputation

```python
data['Bedroom2']
```

```python
from sklearn.impute import SimpleImputer

 

const_imputer = SimpleImputer(strategy = 'constant', fill_value = 3.0)
```

```python
data['Bedroom2'] = const_imputer.fit_transform(data[['Bedroom2']])
```

```python
data.info()
```

## Advance Imputation

### KNN Imputaton

```python
from sklearn.impute import KNNImputer

knn_imputer =  KNNImputer()
```

```python
data['Longtitude'] = knn_imputer.fit_transform(data[['Longtitude']])
```

```python
data.info()
```

### Mice Imputation

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import IterativeImputer

iter_imputer = IterativeImputer()
```

```python
data['Landsize'] = iter_imputer.fit_transform(data[['Landsize']])
```
